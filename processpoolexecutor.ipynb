```python
from concurrent.futures import ProcessPoolExecutor
from time import sleep
```

```python
funct_str = """
from time import sleep

def echo_with_delay(i):
    sleep(10)
    return i
"""
```

```python
with open("myfunctions.py", "w") as f:
    f.write(funct_str)
```

```python
from myfunctions import echo_with_delay
```

# wait=True and cancel_futures=False

```python
exe = ProcessPoolExecutor(max_workers=1)
f1 = exe.submit(echo_with_delay, 1)
f2 = exe.submit(echo_with_delay, 2)
f3 = exe.submit(echo_with_delay, 3)
exe.shutdown(wait=True, cancel_futures=False)
```

```python
f1.done(), f2.done(), f3.done(), f1.cancelled(), f2.cancelled(), f3.cancelled()
```

```python
sleep(12)
```

```python
f1.done(), f2.done(), f3.done(), f1.cancelled(), f2.cancelled(), f3.cancelled()
```
