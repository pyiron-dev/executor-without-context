```python
from executorlib import FluxClusterExecutor
from time import sleep
import shutil
import os
```

```python
def echo_with_delay(i):
    sleep(10)
    return i
```

# wait=True and cancel_futures=False

```python
exe = FluxClusterExecutor(wait=False)
f1 = exe.submit(echo_with_delay, 1)
f2 = exe.submit(echo_with_delay, 2)
f3 = exe.submit(echo_with_delay, 3)
exe.shutdown(wait=True, cancel_futures=False)
```

```python
f1.done(), f2.done(), f3.done(), f1.cancelled(), f2.cancelled(), f3.cancelled()
```

```python
!flux jobs -a
```

```python
sleep(12)
```

```python
!flux jobs -a
```

```python
f1.done(), f2.done(), f3.done(), f1.cancelled(), f2.cancelled(), f3.cancelled()
```

```python
if os.path.exists("executorlib_cache"):
    shutil.rmtree("executorlib_cache")
```
