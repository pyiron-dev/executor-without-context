```python
import shutil
import time
from pathlib import Path
```

```python
from executorlib import FluxClusterExecutor as MyExecutor
```

```python
def slow_step(x):
   time.sleep(4)
   return x + 1
```

```python
cache_dir = Path("test_cache_dir")
cache_dir.mkdir(exist_ok=True)
cache_key = "final"
```

```python
exe = MyExecutor(cache_directory=cache_dir)
```

```python
t0 = time.monotonic()
f1 = exe.submit(slow_step, 10)
f2 = exe.submit(slow_step, x=f1)
f3 = exe.submit(slow_step, x=f2, resource_dict={"cache_key": cache_key})
dt_submit = time.monotonic() - t0
print(f"3x submit() took {dt_submit:.3f}s  (instant — just queued)")
```

```python
t0 = time.monotonic()
exe.shutdown(wait=False, cancel_futures=False)
dt_shutdown = time.monotonic() - t0
print(f"shutdown(wait=False) at {dt_shutdown:.3f}s")
```
